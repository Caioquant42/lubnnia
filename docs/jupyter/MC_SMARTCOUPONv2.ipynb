```python
import whale_wallet as ww
import numpy as np
from datetime import datetime
```

```python
def business_days_between(start_date, end_date):
    # Convert strings to datetime objects if necessary
    if isinstance(start_date, str):
        start_date = datetime.strptime(start_date, '%Y-%m-%d')
    if isinstance(end_date, str):
        end_date = datetime.strptime(end_date, '%Y-%m-%d')
    
    # Use numpy's busday_count to count business days
    business_days = np.busday_count(start_date.date(), end_date.date())
    
    return business_days
```

```python
start = '2024-10-03'
end = '2024-12-27'
print(f"Dias úteis entre {start} -> {end} ==> {business_days_between(start, end)} Dias")
```

```python
#INPUTS
ticker = 'PRIO3' # Ativo
barreira = 38.06 # Barreira de knockout para preço de referência 25,12
ttm = 61 # Tempo até o vencimento em dias utéis.
#S0 = np.array(44.26)
```

```python
period = str(ttm) + 'd'
asset_df = ww.ydata(ticker_symbol=ticker, period=period).get_stock_data()
```

```python
returns_calc = ww.ReturnsCalculator(asset_df)
returns_calc.calculate_returns()
```

```python
S0 = returns_calc.S0
log_returns = returns_calc.log_returns
sorted_values = returns_calc.sorted_values
interpolated_cdf_values = returns_calc.interpolated_cdf_values
```

```python
ww.plot_histogram(log_returns)
```

```python
# Bootstrap method
t_intervals,iterations,additional_iterations = ttm,20000,1
arrivals = ww.Simulation(S0,interpolated_cdf_values,sorted_values)._run_simulation(t_intervals, iterations, additional_iterations)
```

```python
ww.plot_boxplot(S0,arrivals)
```

```python
ww.plot_histogram(arrivals)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Get user input for x
x = barreira
# Calculate the cumulative probability P(X > x)
probability = np.sum(arrivals > x) / len(arrivals)

# Print the cumulative probability
print(f"P(X > {x}) = {probability}")

# Plot histogram
plt.figure(figsize=(10, 6))
plt.hist(arrivals, bins=30, color='skyblue', edgecolor='black', alpha=0.7)
plt.axvline(x, color='red', linestyle='--', linewidth=1.5, label=f"x = {x}")
plt.title('Histogram of Arrivals')
plt.xlabel('Arrival Prices')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```
